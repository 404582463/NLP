```python
import re
from math import log
import copy

tf_matrix = {}
idf_matrix = {}
tf_idf_matrix = {}
output_matrix = {}

documents = []
name_of_docs = []
demo_documents = ['d1.txt','d2.txt','d3.txt','d4.txt','d5.txt']


def get_docs():
    global documents
    global name_of_docs
    
    while True:
        print("Please input the DOCUMENT you want to IMPORT: ")
        print("(1. Input 'over' if no more documents; 2.Input 'demo' to use example documents)\n")
        doc_name = input()
        if doc_name == 'demo':
            documents = demo_documents
            print("Input documents are: ",documents,'\n')
            for i in documents:
                file_name = re.split('\.',i)[0]
                name_of_docs.append(file_name)
            return
        elif doc_name == 'over':
            print("Input documents are: ",documents)
            for i in documents:
                file_name = re.split('\.',i)[0]
                name_of_docs.append(file_name)
            return
        else:
            documents.append(doc_name)
    

def initialize_word_dict(word: str):
    global tf_matrix
    tf_matrix[word] = {}
    
    for j in name_of_docs:
        tf_matrix[word][j] = 0
    
    
def build_tf_matrix():
    global tf_matrix
    
    for i in documents:
        with open(i) as f:
            sentence = f.readline()
            words_in_sentence = sentence.split()
            
        for j in words_in_sentence:
            if j in tf_matrix:
                tf_matrix[j][re.split('\.',i)[0]] += 1
            else:
                initialize_word_dict(j)
                tf_matrix[j][re.split('\.',i)[0]] += 1
    print("The tf matrix is: ")
    for key, value in tf_matrix.items():
        print(key, ' : ', value)
    print('\n')
    
    
def build_idf_matrix():
    global idf_matrix
    idf_matrix = copy.deepcopy(tf_matrix)
    D = len(documents)
    cnt = 0

    for i in idf_matrix:
        for j in tf_matrix[i]:
            if tf_matrix[i][j] != 0:
                cnt += 1
        for j in tf_matrix[i]:
            if tf_matrix[i][j] != 0:
                idf_matrix[i][j] = '{:.2f}'.format(log(D/cnt,2))
        cnt = 0

    print("The idf matrix is: ")
    for key, value in idf_matrix.items():
        print(key, ' : ', value)
    print('\n')

    
def build_tf_idf_matrix():
    global tf_idf_matrix
    tf_idf_matrix = copy.deepcopy(idf_matrix)
    
    for i in tf_idf_matrix:
        for j in tf_idf_matrix[i]:
            if tf_idf_matrix[i][j] != 0:
                tf_idf_matrix[i][j] = '{:.2f}'.format(float(idf_matrix[i][j]) * float(tf_matrix[i][j]))
    print("The tf-idf matrix is: ")
    for key, value in tf_idf_matrix.items():
        print(key, ' : ', value)
    print('\n')
    

# This is just a transpose of tf-idf matrix
def build_output_matrix():
    global output_matrix
    
    for i in name_of_docs:
        output_matrix[i] = {}
    
    for i in tf_idf_matrix:
        for j in tf_idf_matrix[i]:
            output_matrix[j][i] = '{:.2f}'.format(float(tf_idf_matrix[i][j]))
    print("The output matrix is: ")
    for key, value in output_matrix.items():
        print(key, ' : ', value)
    print('\n')

    
def get_tf_idf_value():
    while True:
        print("Please input the DOCUMENT you want to refer to: \n(Input 'over' if you want to end this program)\n")
        i = input()
        if i == 'over':
            print('Program is over')
            return
        print("Please input the WORD you want to refer to: \n(Input 'over' if you want to end this program)\n")
        j = input()
        if j == 'over':
            print('Program is over')
            return
        print("OUTPUT['%s']['%s'] = " %(i,j), output_matrix[i][j],'\n')
    
    
def main():
    get_docs()
    build_tf_matrix()
    build_idf_matrix()
    build_tf_idf_matrix()
    build_output_matrix()
    get_tf_idf_value()

main()
```
